<b>1 часть (выполняем в тренажере):</b>
* преобразуем весь датафрейм в текст;
* потом кодим в base64;
* потом "декодим" (чтобы получить не байты, а текст, т.е. просто убираем "b" (что означает "байты") от полученного ранее результата);
* печатаем полученный "текст" на экране;
* через ctrl+A копируем всё с экрана (в том числе и нужный нам "текст"), заносим в .txt-файл и избавляемся от ненужного - оставляем только строку "текста".<br/>

<b>2 часть (локально в Jupyter Notebook):</b>
* создаем гиперссылку, при нажатии на которую мы скачаем файл с текстом, преобразованным из base64 в csv-формат;
* оборачиваем её в HTML из модуля IPython.display и получаем гиперссылку;
* щелкаем и скачиваем .csv-файл;
* если в файле отображаются непонятные символы, меняем кодировку в Excel ("Data" (Данные) -> "From text" и выберите кодировку utf-8", в следующем окне выбираем "запятую" как разделитель).<br/>
